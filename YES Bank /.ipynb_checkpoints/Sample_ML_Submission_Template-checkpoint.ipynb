# **Project Summary -**

### **Stock Price Prediction Using Machine Learning**

Stock price prediction is a crucial aspect of financial market analysis, enabling investors to make informed decisions based on historical trends and market patterns. This project aimed to develop a machine learning-based system to predict monthly closing stock prices using **Support Vector Regression (SVR), XGBoost, and AutoRegressive Integrated Moving Average (ARIMA)** models. By leveraging historical stock price data, feature engineering techniques, and advanced modeling approaches, we sought to build a robust framework for accurate stock price forecasting.

The project began with data collection and preprocessing, where historical stock price data was obtained and cleaned for further analysis. The dataset included **Open, High, Low, and Close prices**, as well as derived features such as **Moving Averages (Close_MA_3, Close_MA_6) and the Relative Strength Index (RSI)**. Preprocessing steps involved handling missing values using forward-fill and backward-fill techniques, ensuring that no NaN values interfered with model training. Additionally, the data was sorted chronologically to maintain the integrity of time-series forecasting.

Feature engineering played a pivotal role in improving model accuracy. Features such as **average price, stock price range, and moving averages** were incorporated to provide additional insights into market trends. Moving averages, in particular, helped smooth fluctuations and identify long-term trends, while RSI provided valuable signals about stock momentum and potential reversals. These engineered features allowed machine learning models to capture essential patterns and improve predictive performance.

Three different models were implemented to compare their effectiveness in stock price forecasting. **ARIMA**, a statistical model commonly used for time-series forecasting, was applied to identify trends and seasonality in the stock price data. However, it exhibited limitations when dealing with non-linear relationships and complex patterns. **SVR**, a machine learning regression model, was chosen for its ability to map data into higher-dimensional spaces and capture non-linearity effectively. Finally, **XGBoost**, a powerful gradient-boosting algorithm, was employed due to its strong predictive capabilities and ability to handle structured data efficiently.

To evaluate the models, we utilized key performance metrics such as **Root Mean Squared Error (RMSE), Mean Absolute Percentage Error (MAPE), and R² Score**. RMSE measured the average squared difference between actual and predicted stock prices, ensuring that lower values indicated better accuracy. MAPE helped assess the percentage error in predictions, while the R² score determined how well the model explained variance in stock prices. Among the models tested, **SVR emerged as the most effective**, achieving the lowest RMSE of **13.26** and the highest R² score of **0.9828**, outperforming both ARIMA and XGBoost.

To interpret the predictions and understand feature importance, **SHAP (Shapley Additive Explanations)** analysis was conducted. The results revealed that **Moving Averages (Close_MA_3, Close_MA_6) and RSI were the most influential factors** in stock price forecasting. This insight validated the importance of feature engineering in improving model accuracy and provided valuable information for traders and investors.

For practical implementation, the trained **SVR model was saved using Pickle**, enabling seamless deployment for future predictions without retraining. The model was tested on unseen data, and its predictions were validated against actual stock prices, demonstrating its robustness and reliability in real-world applications. Furthermore, future stock prices for the next 12 months were forecasted using the trained SVR model, providing actionable insights for investors.

In conclusion, this project successfully developed a machine learning-based framework for stock price prediction, showcasing the power of **SVR in capturing market trends and patterns**. By leveraging historical data, feature engineering techniques, and model evaluation strategies, we demonstrated that machine learning can provide valuable insights into financial forecasting. Moving forward, further improvements can be made by incorporating additional market indicators, exploring deep learning approaches, and fine-tuning hyperparameters to enhance predictive accuracy. This project highlights the **potential of machine learning in financial analysis and its ability to drive informed investment decisions.**

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code
# Calculate skewness and kurtosis
skewness = skew(df['Close'])
kurt = kurtosis(df['Close'])

# Plot histogram with KDE
plt.figure(figsize=(8, 5))
sns.histplot(df['Close'], bins=30, kde=True, color='blue', alpha=0.6)

# Annotate skewness and kurtosis
plt.axvline(df['Close'].mean(), color='red', linestyle='dashed', label='Mean')
plt.text(df['Close'].mean(), 5, f"Skewness: {skewness:.2f}\nKurtosis: {kurt:.2f}", 
         fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.title('Distribution of Closing Prices with Skewness & Kurtosis')
plt.legend()
plt.show()
````

##### 1. Why did you pick the specific chart?

This histogram was selected to analyze how closing prices have been distributed historically.
Understanding skewness and kurtosis helps in identifying asymmetry in stock price movements.

##### 2. What is/are the insight(s) found from the chart?

Positive skewness (1.25) confirms that most closing prices were low, with fewer instances of higher prices.
Kurtosis (0.45) shows that there were no extreme price spikes, meaning the stock has followed a gradual decline.
The concentration of data at lower price points reflects a prolonged decrease in stock value.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, traders can use this analysis to predict future price distributions and assess whether the stock is undervalued or risky.
Investors looking for entry points can use skewness to determine the likelihood of price recovery.

Yes, since most closing prices remain below ₹100, it suggests long-term price deterioration.
A right-skewed distribution means the stock rarely closes at high values, reducing investor confidence.

````python
# Handling Outliers & Outlier treatments

#Winsorization, preserves stock trends while capping extreme outliers

# Define capping thresholds (1st & 99th percentile)
low_percentile = df[['Open', 'High', 'Low', 'Close']].quantile(0.01)
high_percentile = df[['Open', 'High', 'Low', 'Close']].quantile(0.99)

# Apply Winsorization (Capping Outliers)
df_winsorized = df.copy()
for col in ['Open', 'High', 'Low', 'Close']:
    df_winsorized[col] = np.where(df[col] > high_percentile[col], high_percentile[col], 
                                  np.where(df[col] < low_percentile[col], low_percentile[col], df[col]))

plt.figure(figsize=(10, 5))
sns.boxplot(data=df_winsorized[['Open', 'High', 'Low', 'Close']])
plt.title("Boxplot After Winsorization (Outlier Capping)")
plt.show()

df_winsorized['Volatility'] = (df_winsorized['High'] - df_winsorized['Low']) / df_winsorized['Open']
df_winsorized['Close_MA_6'] = df_winsorized['Close'].rolling(window=6).mean()

df_winsorized['Prev_Close_1'] = df_winsorized['Close'].shift(1)
df_winsorized['Prev_Close_2'] = df_winsorized['Close'].shift(2)
````

### 4. Feature Manipulation and Selection

#### 1. Feature Manipulation

````python
# Manipulate Features to minimize feature correlation and create new features
df['Avg_Price'] = (df['Open'] + df['High'] + df['Low'] + df['Close']) / 4
df['Range'] = df['High'] - df['Low']

df['Prev_Close_1'] = df['Close'].shift(1)  # Last month’s close
df['Prev_Close_3'] = df['Close'].shift(3)  # Three months ago
df['Prev_Close_6'] = df['Close'].shift(6)  # Six months ago

df['Close_MA_3'] = df['Close'].rolling(window=3).mean()
df['Close_MA_6'] = df['Close'].rolling(window=6).mean()

def compute_rsi(data, window=14):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

df['RSI'] = compute_rsi(df['Close'])
df.tail()
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
# Define search space for (p, d, q)
space = {
    'p': hp.choice('p', range(0, 5)),
    'd': hp.choice('d', range(0, 3)),
    'q': hp.choice('q', range(0, 5))
}

# Define the objective function for Bayesian Optimization
def objective(params):
    try:
        model = ARIMA(df['Close'], order=(params['p'], params['d'], params['q']))
        results = model.fit()
        preds = results.predict(start=int(len(df) * 0.8), end=len(df)-1)
        rmse = np.sqrt(mean_squared_error(df['Close'].iloc[int(len(df) * 0.8):], preds))
        return {'loss': rmse, 'status': STATUS_OK}
    except:
        return {'loss': float("inf"), 'status': STATUS_OK}

# Run Bayesian Optimization
trials = Trials()
best_params = fmin(fn=objective, space=space, algo=tpe.suggest, max_evals=50, trials=trials)

# Print best parameters
p_best, d_best, q_best = best_params['p'], best_params['d'], best_params['q']
print(f"\nBest ARIMA Order Found: ({p_best}, {d_best}, {q_best})")


# Fit the Algorithm
# Predict on the model
# Train ARIMA with best hyperparameters
arima_model = ARIMA(df['Close'], order=(p_best, d_best, q_best))
arima_result = arima_model.fit()

# Train-Test Split
train_size = int(len(df) * 0.8)
train, test = df['Close'][:train_size], df['Close'][train_size:]

# Make Predictions
preds_arima = arima_result.predict(start=len(train), end=len(df)-1)

# Evaluate Performance
rmse_arima = np.sqrt(mean_squared_error(test, preds_arima))
print(f"ARIMA({p_best},{d_best},{q_best}) RMSE: {rmse_arima:.4f}")


````

##### Which hyperparameter optimization technique have you used and why?

The hyperparameter optimization technique used here is Bayesian Optimization, implemented using Tree-structured Parzen Estimator (TPE) via the hyperopt library.

Bayesian Optimization was used because:

Efficient Search: Unlike GridSearchCV and RandomSearchCV, Bayesian Optimization intelligently explores the hyperparameter space by balancing exploration (trying new parameters) and exploitation (focusing on promising areas).

Computational Efficiency: Instead of exhaustively searching all combinations like GridSearchCV or randomly sampling like RandomSearchCV, it models the objective function and selects the next set of parameters based on prior results.

Adaptive Learning: It learns from previous evaluations to focus on the most promising hyperparameter values, reducing the number of evaluations needed.

Here, Bayesian Optimization is used to find the optimal (p, d, q) values for ARIMA by minimizing the RMSE on validation data.
